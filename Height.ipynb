```python
import open3d as o3d
import numpy as np
import os

def calculate_tree_height(pcd_file):
    """
    This function calculates the height of a tree from a point cloud file by 
    finding the difference between the maximum and minimum Z-values in the point cloud.

    Parameters:
    pcd_file (str): The path to the point cloud file.

    Returns:
    float: The height of the tree.
    """
    # Load the point cloud
    pcd = o3d.io.read_point_cloud(pcd_file)
    
    # Convert the point cloud to a numpy array
    points = np.asarray(pcd.points)
    
    # Calculate the minimum and maximum Z values (height)
    min_z = np.min(points[:, 2])
    max_z = np.max(points[:, 2])
    
    # Tree height is the difference between the maximum and minimum Z values
    height = max_z - min_z
    return height

def calculate_heights_in_folder(folder_path):
    """
    This function calculates the height of each tree in a folder containing point cloud files.

    Parameters:
    folder_path (str): The path to the folder containing the point cloud files.

    Returns:
    dict: A dictionary where keys are filenames and values are the corresponding tree heights.
    """
    heights = {}
    for filename in os.listdir(folder_path):
        # Check if the file is a point cloud file
        if filename.endswith('.pcd') or filename.endswith('.ply'):
            file_path = os.path.join(folder_path, filename)
            # Calculate the height of the tree in the file
            height = calculate_tree_height(file_path)
            # Store the height in the dictionary
            heights[filename] = height
    return heights

# Specify the folder path containing the point cloud files
folder_path = r"C:\Users\wenru\Desktop\TLS TREE"

# Calculate the heights of the trees in the folder
heights = calculate_heights_in_folder(folder_path)

# Print the calculated heights
for filename, height in heights.items():
    print(f'Tree: {filename}, Height: {height:.2f} meters')
```

```python
import open3d as o3d
import numpy as np
import os

def calculate_tree_height(ply_file):
    """
    This function calculates the height of a tree from a point cloud file (.ply)
    by finding the difference between the maximum and minimum Z-values in the point cloud.

    Parameters:
    ply_file (str): The path to the point cloud (.ply) file.

    Returns:
    float: The height of the tree.
    """
    # Load the point cloud from the .ply file
    pcd = o3d.io.read_point_cloud(ply_file)
    
    # Convert the point cloud to a numpy array
    points = np.asarray(pcd.points)
    
    # Calculate the minimum and maximum Z values (height)
    min_z = np.min(points[:, 2])
    max_z = np.max(points[:, 2])
    
    # Tree height is the difference between the maximum and minimum Z values
    height = max_z - min_z
    return height

def calculate_heights_in_subfolders(folder_path):
    """
    This function calculates the height of trees in each subfolder that contains a point_cloud.ply file.

    Parameters:
    folder_path (str): The root folder containing subfolders with point cloud files.

    Returns:
    dict: A dictionary where the keys are subfolder names and values are the corresponding tree heights.
    """
    heights = {}
    
    # Iterate through subfolders in the root folder
    for subfolder_name in os.listdir(folder_path):
        subfolder_path = os.path.join(folder_path, subfolder_name)
        
        # Ensure that the current path is a directory
        if os.path.isdir(subfolder_path):
            # Construct the path to the point cloud file in the subfolder
            ply_file_path = os.path.join(subfolder_path, 'point_cloud.ply')
            
            # Check if the point_cloud.ply file exists in the subfolder
            if os.path.isfile(ply_file_path):
                # Calculate the tree height using the .ply file
                height = calculate_tree_height(ply_file_path)
                heights[subfolder_name] = height  # Use subfolder name as the key
            else:
                # Warning message if the point_cloud.ply file is missing
                print(f"Warning: No point_cloud.ply file found in {subfolder_path}")
    
    return heights

# Specify the root folder path containing subfolders with point cloud files
folder_path = r"C:\Users\wenru\Desktop\POLYCAM"

# Calculate tree heights in subfolders
heights = calculate_heights_in_subfolders(folder_path)

# Print the calculated heights for each subfolder
for subfolder, height in heights.items():
    print(f'Tree in subfolder: {subfolder}, Height: {height:.2f} meters')
```

```python
import open3d as o3d
import numpy as np
import matplotlib.pyplot as plt

def load_point_cloud(file_path):
    """
    Loads a point cloud file and returns the point data as a NumPy array.

    Parameters:
    file_path (str): The path to the point cloud file.

    Returns:
    np.ndarray: A NumPy array containing the points in the point cloud.
    """
    # Read the point cloud file using Open3D
    point_cloud = o3d.io.read_point_cloud(file_path)
    # Convert the point cloud data to a NumPy array
    return np.asarray(point_cloud.points)

def plot_tree_height(points):
    """
    Plots the tree height based on point cloud data by displaying an xz-plane scatter plot.

    Parameters:
    points (np.ndarray): The point cloud data.
    """
    # Normalize the z-axis by setting the lowest point to z = 0
    points[:, 2] -= np.min(points[:, 2])
    
    # Find the minimum and maximum z-values (the height range of the tree)
    z_min = np.min(points[:, 2])
    z_max = np.max(points[:, 2])
    
    # Create a new figure with a custom size
    plt.figure(figsize=(6, 12))
    
    # Plot all points in the xz-plane as a scatter plot
    plt.scatter(points[:, 0], points[:, 2], c='black', s=1, label='All Points')
    
    # Add horizontal lines marking the lowest and highest points of the tree
    plt.axhline(y=z_min, color='blue', linestyle='--', linewidth=1, label=f'Lowest Point: {z_min:.2f}m')
    plt.axhline(y=z_max, color='green', linestyle='--', linewidth=1, label=f'Highest Point: {z_max:.2f}m')
    
    # Set the labels (removing the x-axis label)
    plt.xlabel('')  # Removing x-axis label
    plt.ylabel('Height (m)')
    plt.legend()
    
    # Set the aspect ratio
    plt.gca().set_aspect('auto', adjustable='box')
    
    # Show the plot
    plt.show()

# Specify the file path to the point cloud
file_path = r"C:\Users\wenru\Desktop\TLS TREE\2023-05-25.27.signing-oak.tree - Cloud.clone.ply"

# Load the point cloud data
points = load_point_cloud(file_path)

# Plot the tree height visualization
plot_tree_height(points)

```
